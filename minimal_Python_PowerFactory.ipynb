External scripts should usually use `GetApplicationExt()` because it ensures a PowerFactory instance is available, while `GetApplication()` is used when connecting to an already running instance (commonly in internal scripts).

```python
import powerfactory as pf # Import powerfactory module
try:
        app = pf.GetApplicationExt() # External Python scripts should use GetApplicationExt()
        if not app:
            raise RuntimeError("Failed to connect to PowerFactory.")
        print("✓ Connected to PowerFactory")
except Exception as e:
        print(f"✗ Error connecting to PowerFactory: {str(e)}")
        raise
```

### STEP2: Activate a PowerFactory project and a study case within the activated project


```python
# PowerFactory Project Configuration
Project_name = 'Example Model'
Study_case_name = 'Study Case'
```

`ActivateProject()` is used to activate project. Python syntax requires use of the dot (.) operator instead of the colon (:) in order to access element parameters of objects (i.e. name, out of service flag, etc). For example, `study_case.Activate`.

```python
# Activate Project
if app.ActivateProject(Project_name) != 0:
    raise ValueError(f"Failed to activate project: {Project_name}")
print(f"✓ Activated Project: {Project_name}")

# Activate Study Case
study_cases = app.GetProjectFolder('study')
study_case = None
for case in study_cases.GetContents():
    if case.loc_name == Study_case_name:
        study_case = case
        break

if not study_case:
    raise ValueError(f"Study case {Study_case_name} not found.")

study_case.Activate()
print(f"✓ Activated Study Case: {Study_case_name}")

# Run powerfactory in interative mode
app.Show() # to check inside PowerFactory GUI if the activation is working properly, otherwise the script will fail here
```
